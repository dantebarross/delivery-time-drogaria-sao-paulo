# Distance matrix

latitude:
- diminuir 0,01 = 1,5km para norte
- aumentar 0,01 = 1,5km para sul

longitude:
- diminuir 0,01 = 1,5km para leste
- aumentar 0,01 = 1,5km para oeste

```python
def new_lat(latitude):
  north_latitude_distances = []
  south_latitude_distances = []
  distance = 0.01

  for distances in range(10):
    north_latitude_distances.append(latitude - distance)
    south_latitude_distances.append(latitude + distance)
    distance += 0.01
  
  return north_latitude_distances, south_latitude_distances


def new_lng(lng):
  east_latitude_distances = []
  west_latitude_distances = []
  distance = 0.01

  for distances in range(10):
    east_latitude_distances.append(lng - distance)
    west_latitude_distances.append(lng + distance)
    distance += 0.01
  
  return east_latitude_distances, west_latitude_distances
```

```python
import ast

def get_lat(row):
  latitude = ast.literal_eval(row['geometry'])['location']['lat']
  return latitude

def get_long(row):
  longitude = ast.literal_eval(row['geometry'])['location']['lng']
  return longitude
```

```python
df['latitude'] = df.apply(get_lat, axis='columns')
df['longitude'] = df.apply(get_long, axis='columns')
```

```python
def get_travel_time(origin_lat, origin_lng, departure_time, destination_lat, destination_lng):
  best_guess_travel = map_client.distance_matrix(
        departure_time=departure_time, origins=(origin_lat, origin_lng), destinations=(destination_lat, destination_lng), 
        mode='driving', traffic_model='best_guess')
    
  pessimistic_travel = map_client.distance_matrix(
        departure_time=departure_time, origins=(origin_lat, origin_lng), destinations=(destination_lat, destination_lng), 
        mode='driving', traffic_model='pessimistic')
  
  return best_guess_travel, pessimistic_travel


def get_travel_time_by_distance(latitude, longitude, departure_time):
  north_latitudes, south_latitudes = new_lat(latitude)
  east_latitudes, west_latitudes = new_lng(longitude)

  north_travel_time = []
  south_travel_time = []
  east_travel_time = []
  west_travel_time = []

  for i in range(10):
    north_travel_time.append(get_travel_time(   latitude, longitude, departure_time, north_latitudes[i], longitude   ))
    south_travel_time.append(get_travel_time(   latitude, longitude, departure_time, south_latitudes[i], longitude   ))
    east_travel_time.append(get_travel_time(   latitude, longitude, departure_time, latitude, east_latitudes[i]   ))
    west_travel_time.append(get_travel_time(   latitude, longitude, departure_time, latitude, west_latitudes[i]   ))
  
  return north_travel_time, south_travel_time, east_travel_time, west_travel_time
```

```python
def get_travel_time_by_distance_by_hour_of_day(dia, original_latitude, original_longitude):
  lista_de_horarios = []

  for i in range(25):
    lista_de_horarios.append(dia)
    dia += timedelta(hours=1)


  travel_time_by_distance_by_day = {}

  for horario in lista_de_horarios:
    n, s, e, w = get_travel_time_by_distance(original_latitude, original_longitude, horario)
    means = get_mean_by_distance(n, s, e, w)
    travel_time_by_distance_by_day.update({(horario - timedelta(hours=3)): means})
  
  return travel_time_by_distance_by_day
```

```python
def get_delivery_time(mean_by_distance_infos, raio):
  values = {'1.5 km': 0, '3.0 km': 1, '4.5 km': 2, '6.0 km': 3, '7.5 km': 4, '9.0 km': 5, '10.5 km': 6, '12.0 km': 7, '13.5 km': 8, '15.0 km': 9}
  mean_by_distance = ast.literal_eval(mean_by_distance_infos)
  mean_by_distance = list(mean_by_distance.values())


  best_guess_mean = sum(d['best_guess'] for d in mean_by_distance[0:values[raio]]) / len(mean_by_distance[0:values[raio]])
  pessimistic_guess_mean = sum(d['pessimistic_guess'] for d in mean_by_distance[0:values[raio]]) / len(mean_by_distance[0:values[raio]])

  
  return best_guess_mean, pessimistic_guess_mean, raio
```
